```python
import pandas as pd
import zipfile
```

```python
zipfilepath = "/content/melb_data.csv"
```

```python
#pd.read use to read the dataset that we have import in the form of exacl sheet
melbourne_data = pd.read_csv(zipfilepath)
#now columns used to separate the exacl sheet to read that how many columns are the sheet that we can use further
melbourne_data.columns
```

```python
#removes the rows that contains NULL values
melbourne_data = melbourne_data.dropna(axis=0)
```

```python
#price is one of the column name in the sheet to read the price columns
y = melbourne_data.Price
```

```python
#features defined the price attributes on which price is dependent
melbosurne_feature=['Rooms','Bathroom','Landsize','Longtitude','Lattitude']
```

```python
#define data featuers in the dataset
x = melbourne_data[melbosurne_feature]
```

```python
#describe used to tell detail about the data features in data
x.describe()
```

```python
#haed used to define the header features
x.head()
```

```python
#imports the DecisionTreeRegressor class from the sklearn.tree module. The DecisionTreeRegressor class is used for building and training decision tree regressor models.
from sklearn.tree import DecisionTreeRegressor
#random state means that the model will produce the same results each time it is trained on the same data.
melbourne_modal = DecisionTreeRegressor(random_state=1)
#fit() method is an essential part of training any machine learning model. It is the process by which the model learns from the data and becomes able to make accurate predictions.
melbourne_modal.fit(x,y)
```

```python
print("making perdication for the following five houses")
print(x.head())
print("the perdication are")
print(melbourne_modal.predict(x.head()))
```
